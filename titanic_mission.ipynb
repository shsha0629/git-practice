3. 생존자(survived=1)만 필터링

```python
titanic[titanic['Survived'] == 1]
```

4. 성별에 따른 생존율 계산

```python
titanic.groupby('Sex')['Survived'].mean()
```

5. age 결측치를 중앙값으로 채우기

```python
df = titanic.copy()
df['Age'] = df['Age'].fillna(df['Age'].median())

df.isnull().sum()
```

---

##### APPLIED | 그룹별 분석

6. 등급별 평균 나이·요금·생존율 (agg 사용)

```python
df.groupby('Pclass').agg(
    생존율=('Survived', 'mean'), 
    요금=('Fare', 'mean'),
    인원=('Survived', 'count')
)
```

7. 1등급 여성 승객 이름 목록

```python
df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]['Name'].tolist()
```

8. 35세 초과 + 1등급 + 남성 승객 수

```python
df[(df['Age'] > 35) & (df['Pclass'] == 1) & (df['Sex'] == 'male')].shape[0]
```

9. 탑승항(embark_town)별 생존율 비교

```python
df.groupby('Embarked')['Survived'].mean()
```

##### CHALLENGE | 쇼핑몰 주문 데이터

```python
import numpy as np

np.random.seed(42)
n = 150
shop = pd.DataFrame({
    'customer': np.random.choice(['김민준','이서현','박지우','최하은','정도윤'], n),
    'category': np.random.choice(['의류','전자기기','식품','생활용품','뷰티'], n),
    'quantity': np.random.randint(1, 6, n),
    'price':    np.random.choice([15000,29000,45000,89000,120000], n),
    'rating':   np.random.choice([1,2,3,4,5], n, p=[0.05,0.10,0.20,0.35,0.30]),
    'returned': np.random.choice([True, False], n, p=[0.1, 0.9])
})
shop.loc[np.random.choice(shop.index, 8, replace=False), 'rating'] = float('nan')
shop.to_csv('shop_orders.csv', index=False)

df = pd.read_csv('shop_orders.csv')
```

Q1. 카테고리별 총 매출 내림차순

```python
df.groupby('category')['price'].mean().round(2).sort_values(ascending=False)
```

Q2. 반품률 가장 높은 카테고리

```python
df.groupby('category')['returned'].mean().sort_values(ascending=False).head(1)
```
